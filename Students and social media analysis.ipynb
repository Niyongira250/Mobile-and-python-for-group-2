```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\ERIC\Downloads\Students Social Media Addiction_with_continent_and_missing.csv")

print(df.head())
print(df.info())
print(df.describe())
# we loaded data from local  storage and tried to understand structure of dataset
```

```python
#checking null values
df.isnull().sum()
```

```python
#we droped empty rows at important columns
df = df.dropna(subset=['Student_ID'])
```

```python
df = df.dropna(subset=['Age', 'Gender', 'Country'])
```

```python
# we supervised dataframe state 

print(df.info())
```

```python
#we also checked null again
print(df.isnull().sum())
```

```python
#filling numeric columns with mean 
numeric_cols = df.select_dtypes(include='number').columns  # selecting  numeric columns in our datasets
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
#we also checked null again    to check info all important columns doesn't have null  and no need to drop duplicates in our case
print(df.isnull().sum())
```

```python
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()
# our age distribution in dataset
```

```python
#gender distribution
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
# alsmost same number of male and female is in our dataset
```

```python
#continent distribution
sns.countplot(x='Continent', data=df)
plt.title("Students by Continent")
plt.xticks(rotation=45)
plt.show()
```

```python
#time spent on social medias
sns.histplot(df['Avg_Daily_Usage_Hours'], kde=True)
plt.title("Daily Social Media Usage Hours")
plt.show()
```

```python
numeric_cols = df.select_dtypes(include='number')

# ploting numeric columns only visualizzing strong/weak connection between variables
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Numerical Variables")
plt.show()
```

```python
platform_counts = df['Most_Used_Platform'].value_counts()

# we visualized the extent different platforms are mostly used 
plt.figure(figsize=(8,8))
plt.pie(platform_counts, labels=platform_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title("Distribution of Most Used Platforms")
plt.axis('equal') 
plt.show()
```

```python
sns.pairplot(df)
plt.show()
# final pair plot of all variables
```
